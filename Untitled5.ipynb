## Feature Engineering for Clustering

### Subtask:
Aggregate the trade data to create relevant features for each country, such as average export and import concentration with their major trading partners. Normalize these engineered features in preparation for PCA and clustering.


**Reasoning**:
I will pivot the `df` DataFrame to create `df_concentrations`, separating export and import values. Then, I will reset the index, rename the pivoted columns, and remove any rows with missing concentration values. Next, I will group by country and calculate the average export and import concentration, storing the result in `df_features`. Finally, I will rename the concentration columns in `df_features` to reflect the average calculation.



```python
df_concentrations = df.pivot_table(index=['Country_Code', 'Country_Name', 'Year'], columns='Series', values='Value').reset_index()

df_concentrations = df_concentrations.rename(columns={'exports': 'Export_Concentration', 'imports': 'Import_Concentration'})

df_concentrations.dropna(subset=['Export_Concentration', 'Import_Concentration'], inplace=True)

df_features = df_concentrations.groupby(['Country_Code', 'Country_Name']).agg({
    'Export_Concentration': 'mean',
    'Import_Concentration': 'mean'
}).reset_index()

df_features = df_features.rename(columns={
    'Export_Concentration': 'Avg_Export_Concentration',
    'Import_Concentration': 'Avg_Import_Concentration'
})

print('First 5 rows of df_features:')
print(df_features.head())

print('\nDataFrame Info of df_features:')
df_features.info()

print('\nMissing values in df_features:')
print(df_features.isnull().sum())
```

**Reasoning**:
The next instruction is to normalize the engineered features. I will import `StandardScaler` from `sklearn.preprocessing`, select the `Avg_Export_Concentration` and `Avg_Import_Concentration` columns, and then apply `StandardScaler` to these columns, storing the scaled values in new columns within `df_features`.



```python
from sklearn.preprocessing import StandardScaler

# Select the columns to be scaled
X = df_features[['Avg_Export_Concentration', 'Avg_Import_Concentration']]

# Create a StandardScaler object
scaler = StandardScaler()

# Fit and transform the data
X_scaled = scaler.fit_transform(X)

# Store the scaled data back into df_features
df_features['Scaled_Avg_Export_Concentration'] = X_scaled[:, 0]
df_features['Scaled_Avg_Import_Concentration'] = X_scaled[:, 1]

print('First 5 rows of df_features after scaling:')
print(df_features.head())

print('\nDataFrame Info of df_features after scaling:')
df_features.info()

print('\nMissing values in df_features after scaling:')
print(df_features.isnull().sum())
```

## PCA for Dimensionality Reduction

### Subtask:
Apply Principal Component Analysis (PCA) to the processed features to reduce their dimensionality to two components. This will be used for visualizing the clusters.


**Reasoning**:
To reduce the dimensionality of the data for visualization and clustering, I will apply Principal Component Analysis (PCA) to the scaled export and import concentration features. This will involve importing PCA, selecting the relevant columns, fitting and transforming the data, and then adding the two principal components back to the DataFrame.



```python
from sklearn.decomposition import PCA

# Select the scaled features for PCA
X_pca = df_features[['Scaled_Avg_Export_Concentration', 'Scaled_Avg_Import_Concentration']]

# Initialize PCA with 2 components
pca = PCA(n_components=2)

# Fit PCA to the scaled data and transform it
principal_components = pca.fit_transform(X_pca)

# Store the principal components in new columns in df_features
df_features['PC1'] = principal_components[:, 0]
df_features['PC2'] = principal_components[:, 1]

print('First 5 rows of df_features after PCA:')
print(df_features.head())

print('\nDataFrame Info of df_features after PCA:')
df_features.info()

print('\nMissing values in df_features after PCA:')
print(df_features.isnull().sum())
```

## K-Means Clustering

### Subtask:
Perform K-Means clustering on the prepared trade features, setting the number of clusters to 4 as requested. Assign each country to its respective cluster.


**Reasoning**:
I will perform K-Means clustering on the scaled export and import concentration features, setting the number of clusters to 4, and then assign the resulting cluster labels back to the DataFrame.



```python
from sklearn.cluster import KMeans

# Select the scaled features for clustering
X_cluster = df_features[['Scaled_Avg_Export_Concentration', 'Scaled_Avg_Import_Concentration']]

# Initialize KMeans with 4 clusters and a random state for reproducibility
kmeans = KMeans(n_clusters=4, random_state=42, n_init='auto')

# Fit the KMeans model to the scaled data
kmeans.fit(X_cluster)

# Predict the cluster labels and add them to df_features
df_features['Cluster'] = kmeans.labels_

print('First 5 rows of df_features after clustering:')
print(df_features.head())

print('\nDataFrame Info of df_features after clustering:')
df_features.info()

print('\nMissing values in df_features after clustering:')
print(df_features.isnull().sum())
```

## Visualize Clusters

Generate a scatter plot using the first two PCA components, with each point representing a country and colored according to its assigned cluster. Additionally, create a histogram to visualize the distribution of countries across the four clusters.


**Reasoning**:
To visualize the clusters, I will generate a scatter plot of the first two PCA components, coloring points by cluster. Additionally, I will create a histogram to show the distribution of countries across the four clusters, as requested in the subtask.

*   **Cluster 2: Region dependency**
